## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests
import json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = json.loads(r.text) #用json把str轉成list
print('data_pieces=',len(response))
for res in response:
    print('data_column=',len(res))
    for r in res:
        print(r)
    print('----------------------------------')
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for res in response:
    print('標題:',res['title'])
    print('貼文時間:',res['createdAt'])
    #print(res['updatedAt']) 貼文更新時間
    print('留言人數:',res['commentCount'])
    print('按讚人數:',res['likeCount'])
    print('-----------------------------------------')
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

popular = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
notpopular = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
pop = json.loads(popular.text)
notpop = json.loads(notpopular.text) 

countComments=0
countLikes=0
sizeOfPop = len(pop)
print('Popular Articles:',sizeOfPop)
for p in pop:
    countComments += p['commentCount']
    countLikes += p['likeCount']
print('Popular Articles Average Comments=%.1f, Average Likes=%.1f'%(countComments/sizeOfPop, countLikes/sizeOfPop))

countComments=0
countLikes=0
sizeOfNotpop = len(notpop)
print('-------------------------------------------------------')
print('Not Popular Articles:',sizeOfNotpop)
for notp in notpop:
    countComments += notp['commentCount']
    countLikes += notp['likeCount']
print('Not Popular Articles Average Comments=%.1f, Average Likes=%.1f'%(countComments/sizeOfNotpop, countLikes/sizeOfNotpop))
```
