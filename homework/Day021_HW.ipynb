## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
from datetime import datetime,timedelta

browser = webdriver.Chrome(executable_path ='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
def eraseZero(y,m,d):
    y = str(int(y))
    m = str(int(m))
    d = str(int(d))
    return y,m,d
```

```python
#取得今天日期
now = datetime.now()
nowDate = now.strftime('%Y/%m/%d') # (日期型態).strftime('格式')
print(nowDate)
year,mon,day= nowDate.split('/') #str型態 但是有0

year,mon,day = eraseZero(year,mon,day)
#為了把 01 08 個位數的0去除
print(year,mon,day)
```

```python
#選擇日期
def selectDate(year,mon,day):
    selectDay = Select(browser.find_element_by_id("selD"))
    selectMonth = Select(browser.find_element_by_id("selM"))
    selectYear = Select(browser.find_element_by_id("selY"))

    selectYear.select_by_value(year)
    selectMonth.select_by_value(mon)
    selectDay.select_by_value(day)

    #點擊搜尋
    browser.find_element_by_id('button').click()
```

## 2. 取出現在時間兩小時內的新聞

```python
#先下拉網頁條
import time
def drawScroll():
    time.sleep(2)
    browser.execute_script("window.scrollTo(0,"+str(10000*i)+");") #優化版本
    #time.sleep(1)
    #browser.execute_script("window.scrollTo(0, 10000);") #scroll值會停在10000的位置
```

```python
#選擇日期
selectDate(year,mon,day)
```

```python
#下拉讀取條
drawScroll()
#抓取搜尋後的新聞資料
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html.parser')
newsList = soup.find('div',class_='part_list_2')
```

```python
#找兩個小時內的時間 新聞
from datetime import datetime,timedelta

now = datetime.now()
nowDate = now.strftime('%Y/%m/%d') #今天日期

for news in newsList:
    if(news.find('span')==-1):continue
    else: 
        newsTime = news.find('span').text
        if(newsTime.find(nowDate)==0): #今天日期確定
            newsTime_Date = datetime.strptime(newsTime, '%Y/%m/%d %H:%M') #字串轉成時間型態格式
            if( now - newsTime_Date < timedelta(hours=2)): #日期時間比較 是否在兩小時內
                print(newsTime,news.find('a').text) #用.string會因為影片icon顯示none
```

```python
browser.quit()
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
browser = webdriver.Chrome(executable_path ='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
now = datetime.now()
day3Ago = now - timedelta(days=3) #算三天前的日期

day3Ago_str = day3Ago.strftime("%Y/%m/%d")
year, mon, day3 = day3Ago_str.split('/')
year, mon, day3 = eraseZero(year, mon, day3)
searchTime = datetime( int(year), int(mon), int(day3),15,0)
print(searchTime)
selectDate(year, mon, day3)
```

```python
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0,"+str(10000*i)+");") #scroll值會停在10000的位置
```

```python
resource = browser.page_source
soup = BeautifulSoup(resource, 'html5lib')
newsList = soup.find('div',class_='part_list_2')
#print(newsList)
```

```python
#print(newsList)
for news in newsList:
    if(news.find('span')==-1):continue
    else: 
        newsTime = news.find('span').string
        if(newsTime.find(day3Ago_str)==0): #今天日期確定
            newsTime_Date = datetime.strptime(newsTime, '%Y/%m/%d %H:%M') #字串轉成時間型態格式
            deltaTime = (newsTime_Date - searchTime )
            #print(deltaTime)
            if(deltaTime < timedelta(hours=2) and deltaTime >=  timedelta(hours=0)): #日期時間比較 是否在15~17點
                print(newsTime,news.find('a').text) #用.string會因為影片icon顯示none
                
#日期格式比較特別 若是負的 會變成 負的(更高單位) 加上 原本單位  
#不會直接顯示 負的 原本單位 時間差   (時間差還是相同的)
```

```python
browser.quit()
```
