### ① 取出今天所有的發文

```python
import requests as req
from bs4 import BeautifulSoup as BS
```

```python
url = 'https://www.ettoday.net/news/news-list.htm'
needHeader = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:71.0) Gecko/20100101 Firefox/71.0'}
response = req.get(url,headers=needHeader)

soup = BS(response.text, "html5lib")
data = soup.find(class_="part_list_2").find_all('h3')
num=0
for d in data:
    date = d.find(class_="date").text #不能用class="date"會衝突變數 所以用class_
    title = d.find_all('a')[0].text
    #print(len(d.find_all('a')))
    print(date,title)
    num+=1
print('-------------------')
print(num," pieces data")
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
#自己建立的分類函式
def groupBy(text,group,count):
    for i in range(len(group)):
        if(text == group[i]):  #找到類別符合則 計數+1
            count[i]+=1
            return group,count
    group.append(text) #沒有符合的就建立一個新的類別
    count.append(1)
    return group,count
```

```python
group=[]
count=[]
for d in data:
    category = d.find_all('em')[0].text
    group,count=groupBy(category,group,count)
    #print(
total=0
for i in range(len(group)):
    print(group[i],':',count[i])
    total += count[i]
print(total,' pieces data')
```

### ③ 哪一個類別的文章最多

```python
max_index = count.index(max(count))
print(group[max_index],":",count[max_index])
```
